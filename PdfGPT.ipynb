```python
import openai
import os
import PyPDF2
```

```python
import PyPDF2  # Import the PyPDF2 library for PDF functionality

with open('/content/JavaPreparationKit.pdf', 'rb') as pdf_file:  # Open PDF in binary mode
    pdf_reader = PyPDF2.PdfReader(pdf_file)  # Create PdfReader object

    raw_text = ''
    for i, page in enumerate(pdf_reader.pages):  # Iterate through pages
        content = page.extract_text()  # Extract text from each page
        if content:  # Check for content before concatenation
            raw_text += content

print(raw_text)  # Print the extracted text



```

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(
    separator ="\n",
    chunk_size =800,
    chunk_overlap=200,
    length_function=len
)
texts = text_splitter.split_text(raw_text)
```

```python
from langchain.embeddings.openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(openai_api_key=userdata.get('chatgpt_api_key'))

```

```python
from langchain.vectorstores import FAISS
document_search = FAISS.from_texts(texts, embeddings)
```

```python
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI
```

```python
chain = load_qa_chain(OpenAI(openai_api_key=userdata.get('chatgpt_api_key')), chain_type="stuff")
```

```python
query = "Randomly pick a topic and explain in a very simple manner in 1 lines"
docs = document_search.similarity_search(query)
chain.run(input_documents = docs, question = query)
```
